### Redundancy

#### Expert:

```python
# Step 1: Find all duplicates based on Expert + Price + Brand + Color
duplicate_rows = integrated.duplicated(subset=['Expert', 'Expert Price', 'Expert Brand', 'Expert Color'], keep=False)
duplicates_df = integrated[duplicate_rows].copy()

# Step 2: Group by Expert + Price + Brand + Color → count how many unique categories exist per group
category_counts = duplicates_df.groupby(
    ['Expert', 'Expert Price', 'Expert Brand', 'Expert Color']
)['Expert Category'].nunique().reset_index(name='Num_Categories')

# Step 3: Keep only groups where there is more than 1 unique category → these are the problematic ones
multi_category = category_counts[category_counts['Num_Categories'] > 1]

# Step 4: Reconstruct the final dataframe with only the duplicate rows having different categories
final_duplicates = duplicates_df.merge(
    multi_category[['Expert', 'Expert Price', 'Expert Brand', 'Expert Color']],
    on=['Expert', 'Expert Price', 'Expert Brand', 'Expert Color'],
    how='inner'
)

# Step 5: Print result
print(f"Found {final_duplicates.shape[0]} duplicated rows with DIFFERENT categories.")

# Display the most useful columns for manual review
final_duplicates[['Expert', 'Expert Price', 'Expert Brand', 'Expert Color', 'Expert Category', 'Unieuro Category', 'Euronics Category']].sort_values(by='Expert')
```

```python
# Step 6: Remove problematic duplicates from the main dataframe

# Create a key to easily identify the groups to remove
multi_category['remove_key'] = multi_category['Expert'] + '||' + multi_category['Expert Price'].astype(str) + '||' + multi_category['Expert Brand'] + '||' + multi_category['Expert Color']

# Now create the same key in the original dataframe
integrated['remove_key'] = integrated['Expert'] + '||' + integrated['Expert Price'].astype(str) + '||' + integrated['Expert Brand'] + '||' + integrated['Expert Color']

# Filter out the problematic rows
integrated_no_duplicates = integrated[~integrated['remove_key'].isin(multi_category['remove_key'])].copy()

# Drop the temporary column
integrated_no_duplicates.drop(columns=['remove_key'], inplace=True)

# Print result
print(f"Original dataframe had {integrated.shape[0]} rows.")
print(f"Cleaned dataframe has {integrated_no_duplicates.shape[0]} rows. {integrated.shape[0] - integrated_no_duplicates.shape[0]} rows were removed.")
integrated_no_duplicates
```

```python
integrated_no_duplicates.to_csv("Integrated_No_Duplicates.csv", index=False)
```
