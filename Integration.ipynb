# Expert VS Unieuro

Matching Expert and Unieuro Products using **TF-IDF** and **Cosine Similarity** with Brand and Color Constraints

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from difflib import SequenceMatcher

# Load CSV
Expert = pd.read_csv("Expert_Preprocessed.csv")
Unieuro = pd.read_csv("Unieuro_Preprocessed.csv")
Euronics = pd.read_csv("Euronics_Preprocessed.csv")

# Cleanup function + token sort
def clean_and_sort(text):
    text = str(text).lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)  # rimuove punteggiatura
    tokens = text.split()
    tokens.sort()
    return " ".join(tokens)

# Apply cleanup to product names
Expert["Clean_Product"] = Expert["Product"].apply(clean_and_sort)
Unieuro["Clean_Product"] = Unieuro["Product"].apply(clean_and_sort)
Euronics["Clean_Product"] = Euronics["Product"].apply(clean_and_sort)

# Prepare lists for TF-IDF
products_expert = Expert["Clean_Product"].tolist()
products_unieuro = Unieuro["Clean_Product"].tolist()
combined = products_expert + products_unieuro

# Calculate TF-IDF and cosine similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(combined)
similarity_matrix = cosine_similarity(tfidf_matrix)

# Matching with threshold ≥ 0.7
len_ex = len(products_expert)
matches = []

for i in range(len_ex):
    candidate_matches = []
    for j in range(len(products_unieuro)):
        sim = similarity_matrix[i][len_ex + j]
        if (
            sim >= 0.7 and
            Expert.loc[i, "Color"] == Unieuro.loc[j, "Color"] and
            Expert.loc[i, "Brand"] == Unieuro.loc[j, "Brand"]
        ):
            candidate_matches.append((sim, j))

    if candidate_matches:
        best_sim, best_j = max(candidate_matches, key=lambda x: x[0])
        matches.append({
            "Expert": Expert.loc[i, "Product"],
            "Unieuro": Unieuro.loc[best_j, "Product"],
            "Similarity": best_sim,
            "Expert Price": Expert.loc[i, "Price (€)"],
            "Unieuro Price": Unieuro.loc[best_j, "Price (€)"],
            "Expert Category": Expert.loc[i, "Category"],
            "Unieuro Category": Unieuro.loc[best_j, "Category"],
            "Expert Brand": Expert.loc[i, "Brand"],
            "Unieuro Brand": Unieuro.loc[best_j, "Brand"],
            "Expert Color": Expert.loc[i, "Color"],
            "Unieuro Color": Unieuro.loc[best_j, "Color"]
        })
    else:
        matches.append({
            "Expert": Expert.loc[i, "Product"],
            "Unieuro": np.nan,
            "Similarity": np.nan,
            "Expert Price": Expert.loc[i, "Price (€)"],
            "Unieuro Price": np.nan,
            "Expert Category": Expert.loc[i, "Category"],
            "Unieuro Category": np.nan,
            "Expert Brand": Expert.loc[i, "Brand"],
            "Unieuro Brand": np.nan,
            "Expert Color": Expert.loc[i, "Color"],
            "Unieuro Color": np.nan
        })

# Output DataFrame
output_exe_uni = pd.DataFrame(matches)
#output_exe_uni.to_csv("Expert_Unieuro_Matches.csv", index=False)
```

# Merging

```python
# Base: Expert–Unieuro
final = output_exe_uni.copy()

# Aggiungi colonne da Unieuro–Euronics se Unieuro è presente
for i in range(len(final)):
    elem = final.loc[i, "Unieuro"]

    if elem in output_uni_eur["Unieuro"].tolist():
        idx = output_uni_eur["Unieuro"].tolist().index(elem)

        final.loc[i, "Euronics"] = output_uni_eur.loc[idx, "Euronics"]
        final.loc[i, "Euronics Price"] = output_uni_eur.loc[idx, "Euronics Price"]
        final.loc[i, "Euronics Category"] = output_uni_eur.loc[idx, "Euronics Category"]
        final.loc[i, "Euronics Brand"] = output_uni_eur.loc[idx, "Euronics Brand"]
        final.loc[i, "Euronics Color"] = output_uni_eur.loc[idx, "Euronics Color"]
    else:
        final.loc[i, "Euronics"] = np.nan
        final.loc[i, "Euronics Price"] = np.nan
        final.loc[i, "Euronics Category"] = np.nan
        final.loc[i, "Euronics Brand"] = np.nan
        final.loc[i, "Euronics Color"] = np.nan

# Aggiungi colonne da Expert–Euronics se Expert è presente ma Euronics non lo è ancora
for i in range(len(final)):
    if pd.isna(final.loc[i, "Euronics"]):
        elem = final.loc[i, "Expert"]
        if elem in output_exe_eur["Expert"].tolist():
            idx = output_exe_eur["Expert"].tolist().index(elem)
            final.loc[i, "Euronics"] = output_exe_eur.loc[idx, "Euronics"]
            final.loc[i, "Euronics Price"] = output_exe_eur.loc[idx, "Euronics Price"]
            final.loc[i, "Euronics Category"] = output_exe_eur.loc[idx, "Euronics Category"]
            final.loc[i, "Euronics Brand"] = output_exe_eur.loc[idx, "Euronics Brand"]
            final.loc[i, "Euronics Color"] = output_exe_eur.loc[idx, "Euronics Color"]

# Ordina le colonne finali
cols = [
    'Expert', 'Unieuro', 'Euronics',
    'Expert Price', 'Unieuro Price', 'Euronics Price',
    'Expert Category', 'Unieuro Category', 'Euronics Category',
    'Expert Brand', 'Unieuro Brand', 'Euronics Brand',
    'Expert Color', 'Unieuro Color', 'Euronics Color'
]
final = final[cols]

# Etichetta il tipo di match
final["Match Type"] = final.apply(
    lambda row: (
        "3-way match" if pd.notnull(row["Expert"]) and pd.notnull(row["Unieuro"]) and pd.notnull(row["Euronics"])
        else "2-way match" if sum(pd.notnull(row[c]) for c in ["Expert", "Unieuro", "Euronics"]) == 2
        else "1-way only"
    ), axis=1
)

# (Opzionale) Estrai solo i match completi
#match_3vie = final[final["Match Type"] == "3-way match"]

# Salvataggi (facoltativi)
#final.to_csv("Expert_Unieuro_Euronics_ALL_MATCHES.csv", index=False)
#match_3vie.to_csv("Expert_Unieuro_Euronics_ONLY_3WAY.csv", index=False)

# Anteprima
print(final["Match Type"].value_counts())
final
```
